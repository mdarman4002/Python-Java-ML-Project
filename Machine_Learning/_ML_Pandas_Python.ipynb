```python
pd.merge(df4,df3, on = 'ID', how = 'inner')
```

```python
pd.merge(df4,df3, on = 'ID', how = 'left') # left joint
```

```python
pd.merge(df4,df3, on = 'ID', how = 'right')
```

```python
pd.merge(df4,df3, on = 'ID', how = 'outer')
```

```python
pd.merge(df4,df3, on = 'ID', how = 'outer', indicator = True)
```
